```python
# Applications
# 4.4.1 Random Draws

ts_length = 100 # set length of time series
e_values = [] #empty list

for i in range(ts_length):
    e = np.random.randn()
    e_values.append(e)

plt.plot(e_values)
plt.show()
```

```python
# break this program into parts by generating own function
def generate_data(n):
    e_values = []
    for i in range(n):
        e = np.random.randn()
        e_values.append(e)
    return e_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
# 4.4.2 Adding Conditions to the generate data function
# function now draws from a distribution that is entered

def generate_data(n, generator_type):
    e_values = []
    for i in range(n):
        if generator_type == 'uniform':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        e_values.append(e)
    return e_values

data = generate_data(100, generator_type='uniform')
plt.plot(data)
plt.show()
```

```python
# get rid of conditional by allowing function to take a function for the distribution
def generate_data(n, generator_type):
    e_values = []
    for i in range(n):
        e = generator_type()
        e_values.append(e)
    return e_values

data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```

```python
# 4.5 Recursive Function Calls
# Ex $\x_{t+1} = 2*t_x, x_0 = 1
```

For example $$x_{t+1} = 2*t_x, x_0 = 1$$

```python
# computing with a loop

def x_loop(t):
    x = 1
    for i in range(t):
        x = 2*x
    return x

x_loop(5)
```

```python
# use a recursive solution
def x(t):
    if t == 0:
        return 1
    else:
        return 2*x(t-1)

x(5)
```

```python
# 4.6 Exercises
# Exercise 4.1 write a function for factorial

def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(5))
print(factorial(4))
```

```python
# Exercise 4.2 generate binomial draw

from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        u = uniform()
        if u < p:
            count += 1
    return count

binomial_rv(10, 0.5)
```

```python
# Exercise 4.3
# write a function that returns a realization from the following:
#   1. Flip an unbiased coin 10 times
#   2. if heads occurs k or more times consecutively within this sequence at least once, pay one dollar
#   3. If not pay nothing

from numpy.random import uniform

def draw(k, n, p):

    payoff = 0
    count = 0

    for i in range(n):
        u = uniform()
        count = count + 1 if u < p else 0
        print(count)
        if count == k:
            payoff = 1
    return "Payoff:" + str(payoff)

print(draw(3, 10, 0.5))

# write another function that does the same, but second rule is if head occurs k or more times within this sequence, pay one dollar

def draw(k, n, p):

    payoff = 0
    count = 0

    for i in range(n):
        u = uniform()
        count = count + 1 if u < p else count
        print(count)
        if count == k:
            payoff = 1
    return "Payoff:" + str(payoff)

print(draw(3, 10, 0.5))
```

```python
# Exercise 4.4 Recursive Function
# function to recursively compute Fibonacci numbers

def x(t):
    if t == 0:
        return 0
    if t == 1:
        return 1
    else:
        return x(t-1) + x(t-2)

x(10)
```
